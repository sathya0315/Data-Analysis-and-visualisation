```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.datasets import load_breast_cancer
df=pd.read_csv("/content/2.2 dataset breast cancer (1).csv")
```

```python
df.info()
```

```python
breast=load_breast_cancer()
breast_data=breast.data
print(breast_data)
print(breast_data.shape)
```

```python
breast_labels=breast.target
print(breast_labels)
print(breast_labels.shape)
```

```python
labels=np.reshape(breast_labels,(569,1))
final_breast_data=np.concatenate([breast_data,labels],axis=1)
print(final_breast_data.shape)
```

```python
breast_dataset=pd.DataFrame(final_breast_data)
print(breast_dataset.head())
```

PRINT THE FEATURES THAT ARE THERE IN BREAST CANCER DATASET

```python
features=breast.feature_names
print(features)
```

HERE THE LABEL FIELD IS MISSING SO ADD IT

```python
features_labels=np.append(features,'label')
```

EMBEDDING THE COLUMN NAMES TO THE DATAFRAME

```python
breast_dataset.columns=features_labels
breast_dataset.head()
```

REPLACE FUNCTION

```python
breast_dataset['label'].replace(0,'Benign',inplace=True)
breast_dataset['label'].replace(0,'Malignant',inplace=True)
breast_dataset.tail()
```

NORMALIZING THE FEATURES

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:,features].values
x = StandardScaler().fit_transform(x)
print(x.shape)
```

```python
np.mean(x),np.std(x)
```

CONVERTING NORMALIZED FEATURES INTO TABULAR FORMAT

```python
feat_cols=['feature'+str(i) for i in range(x.shape[1])]
normalized_breast=pd.DataFrame(x,columns=feat_cols)
print(normalized_breast)
```

```python
normalized_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast=PCA(n_components=2)
principalComponents_breast=pca_breast.fit_transform(x)
principal_breast_Df=pd.DataFrame(data=principalComponents_breast,columns=['principal component 1','principal component 2'])
principal_breast_Df.tail()
```

PLOT THE PCA

```python
import matplotlib.pyplot as plt
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.xlabel('Principal Component-1',fontsize=20)
plt.ylabel('Principal Component-1',fontsize=20)
plt.title("Principal Component Analysis of abreast Cancer Dataset",fontsize=20)
targets=['Benign','Malignant']
colore=['r','g']
for target,color in zip(targets,colore):
  indicesToKeep=breast_dataset['label']==target
  plt.scatter(principal_breast_Df.loc[indicesToKeep,'principal component 1'],principal_breast_Df.loc[indicesToKeep,'principal component 2'],c=color,s=5)
  plt.legend(targets,prop={'size':15})
```
